For my data set, I decided to work with the Police Incident Blotter Data from the past 30 days. The data lists the date of the incident, the time of day, where the incident happened, what incident actually occurred, and the offenses given among other things. After looking at the data, I chose to determine the best neighborhood based on how little police incidents occured in the last 30 days. I chose this metric because it works well with arrest data, which gives a better idea of the crime rate going on in each neighborhood. This also helps when considering our overall metric, which is the security of each neighborhood.

To start off, I import everything necessary for the data visualization. I also change the font size and theme of the graphs for later.

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size': 5})
plt.rcParams['figure.figsize'] = [13, 5]
plt.style.use('Solarize_Light2')
```

Next, I import the file containing all of the police incident data for the last 30 days. I then print it out to see what it looks like.

```python
data = pd.read_csv("Incident-Data-Thirty-Days.csv", parse_dates=True)
data
```

This is what the base table looks like, but not everything here is necessary to find out what the best neighborhood is. The main thing we need to look for is how many incidents occur in
each neighborhood. So, I sort the table by the category "INCIDENTNEIGHBORHOOD" to see what I am looking for.

```python
data.groupby("INCIDENTNEIGHBORHOOD").count()
```

By sorting by the category "INCIDENTNEIGHBORHOOD", I realize that pandas sorts the table based on how many incidents occurred in each neighborhood. Using this,
I can sort the data using value_counts to rank the neighborhoods on how many incidents occurred in each neighborhood. By using ascending=True, it sorts the neighborhoods
from least to greatest.

```python
chart = data["INCIDENTNEIGHBORHOOD"].value_counts().sort_values(ascending=True)
chart
```

Now that I have the data sorted the way I want, we need to vizualize it as a graph, making sure the labels are at the proper orientation.

```python
chart.plot.bar(rot = 0)
```

This graph is way too big! You can't even see the labels! For the sake of actually being able to read the graph, I decided to use the top fifteen entries of the data table.

```python
detailed = chart.head(15)
detailed.plot.bar(rot = 0)
```

Using the data from the graph, you can see that there are quite a few neighborhoods that have had little incidents occurr. So, for the sake of ranking the best neighborhoods, I've decided to give each neighborhood with 2 incidents a score of 9/10, those neighborhoods being Summer Hill, Mount Oliver, Fairywood, Esplen, 
Stanton Heights, Oakwood, and Arlington Heights, and each neighborhood with only 1 incident 10/10, those being Duquesne Heights, Glen Hazel, West End, Swissheim Park, and St. Clair.
